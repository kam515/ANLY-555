```python
# Implement a ToyData class with member attributes _data of type list and _size which is an integer

# ToyData will have two member methods plus a constructor
# generate() will randomly generate _size floating point values between 0 and 1 and store them into _data
# display() will display a figure in a histogram of the values in _data
# a constructor which takes one parameter and uses this parameter to initialize the value of size

# implement a main method to rigorously test your new class!!
```

```python
import random
```

```python
# Instantiate the class object
class ToyData:
    """ToyData ..."""
    
    def __init__(self, size):
        """Create a new ToyData instance."""
        self._size = size
        self._data = []
        return
    
    def generate(self):
        """Generate _size floating point values between 0 and 1 and store them into _data"""
        self._data = [random.random() for i in range(0, self._size)]
    
    def display(self):
        """Display a histogram of the data"""
        import matplotlib.pyplot as plt
        hist = plt.hist(self._data)
        return plt.show(hist)
        
```

```python
def main():
    """Testing the ToyData class functionality"""
    for i in range(20):
        size = 15
        toy = ToyData(size)
        toy.generate()
        toy.display()

main()
```
